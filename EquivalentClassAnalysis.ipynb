```python
#https://csr.lanl.gov/data/2017/
file = 'netflow_day-11'
df = pd.read_csv(file, header=None)
df.columns = ['Time','Duration','SrcDevice','DstDevice','Protocol','SrcPort','DstPort','SrcPackets','DstPackets','SrcBytes','DstBytes']
df
```

```python
print('ファイル中のレコード数 : {0}'.format(len(df)))
```

```python
df_device = pd.read_csv(file+'_device.csv', header=None)
print('和集合デバイス数 : {0}'.format(len(df_device)))
```

```python
df_device_intersection = pd.read_csv(file+'_device_intersection.csv', header=None)
print('積集合デバイス数 : {0}'.format(len(df_device_intersection)))
```

```python
device = 'ActiveDirectory' #最大同値類クラスでの中心性デバイス
df_dst = df[df['DstDevice'] == device] #中心性デバイスが通信先
print('中心性デバイスが送信先の回数 : {0}'.format(len(df_dst)))
df_src = df[df['SrcDevice'] == device] #中心性デバイスが通信元
print('中心性デバイスが送信元の回数 : {0}'.format(len(df_src)))
df_communication = df[(df['SrcDevice'] == device) | (df['DstDevice'] == device)] #中心性デバイス通信
print('中心性デバイスの通信回数 : {0}'.format(len(df_communication)))
```

```python
print(df_dst['DstPort'].value_counts()) #中心性デバイスの通信内容(dstで評価)
```

```python
transition = pd.read_csv('transition_count_non0_rate_maxclass.csv', index_col=0) #最大同値類クラスの推移確率
transition
```

```python
maxdevice_row = transition.loc[device] #中心デバイスからの送信先
maxdevice_row_list = maxdevice_row.index.tolist()
maxdevice_row_value = maxdevice_row.tolist()
zip_lists = zip(maxdevice_row_value, maxdevice_row_list)
zip_sort = sorted(zip_lists, reverse=True)
val_sort, dev_sort = zip(*zip_sort)
idx = 1
for dev, val in zip(dev_sort, val_sort):
    print('順位 : ', idx)
    print('中心性デバイスからの送信先', dev)
    print('中心性デバイスからの送信確率', val)
    idx += 1
    if idx > 4:
        break
```

```python
maxdevice_col = transition[device] #中心性デバイスへの送信元
maxdevice_col_list = maxdevice_col.index.tolist()
maxdevice_col_value = maxdevice_col.tolist()
zip_lists = zip(maxdevice_col_value, maxdevice_col_list)
zip_sort = sorted(zip_lists, reverse=True)
val_sort, dev_sort = zip(*zip_sort)
idx = 1
for dev, val in zip(dev_sort, val_sort):
    print('順位 : ', idx)
    print('中心性デバイスへの送信元', dev)
    print('中心性デバイスへの送信確率', val)
    idx += 1
    if idx > 4:
        break
```
